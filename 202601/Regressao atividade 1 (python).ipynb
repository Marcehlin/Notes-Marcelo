Análise de Regressão --- Atividade 1
Enunciado:

Um pesquisador deseja estudar a relação entre o Consumo (km/l) de um automóvel e a sua
Velocidade Média em um percurso fixo.


1. Gere um conjunto de dados coerente com este problema e apresente o código
utilizado.

```python
# definindo coisas e descrevendo os passos
# X: variável regressora; velocidade média do veículo (km/h)
# Y: variável resposta; consumo de combustível (km/l)
# Gerar X's aleatórios e 'Calcular' os Y's tais que Y = beta_0 + beta_1*f(X) + um_certo_erro
# Como o automóvel não é especificado (não sei a marca nem modelo), vou escutar as vozes da minha cabeça e dizer que f(X) é igual a X,
# e os valores para beta_0 e beta_1? vou inventando 
# mesmo que isso pode não ser condizente com a realidade de qualquer automóvel do mundo

import numpy as np
import matplotlib.pyplot as plt

n = 30 # número de observações que queremos gerar
beta_0 = 5
beta_1 = 0.1
X = np.random.default_rng().uniform(0,150,size = n) #gerar uniformemente n números entre 0 a 150
Erros = np.random.normal(0,4,size = n) # uns erros aleatórios,homocedásticos, que seguem uma distribuição normal de media 0 e variancia 4
Y = beta_0 + beta_1*X + Erros # "Calcular" Y
dados = np.column_stack((Y, X)) #juntar os dados
```

2. Ajuste a reta de regressão linear.


```python
plt.scatter(X, Y)
beta_1_chapeu, beta_0_chapeu  = np.polyfit(X, Y, 1)
plt.plot(X, beta_0_chapeu + beta_1_chapeu*X, color='red', label=f' Reta de regressão linear simples: Y = {beta_0_chapeu:.2f} + {beta_1_chapeu:.2f}X')
plt.xlabel('Velocidade Média')
plt.ylabel('Consumo')
```
